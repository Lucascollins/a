```python
#Importando bibliotecas
from selenium import webdriver
from tkinter import *
import pandas as pd
import pyautogui
import time


#Layout
root = tk.Tk()
root.geometry('500x600')
root.resizable(width=0, height=0)
root.title("Bot mandar Mensagem")

#Labels
_nameSheet= Label(root,text ="Nome Planilha  :  ")
_nameSheet.grid(column = 0,row = 0)
_inputSheet = Entry(root,width = 50)
_inputSheet.grid(column = 1,row = 0)

text_mensagem = Label(root,text ="Mensagem  :  ")
text_mensagem.grid(column = 0,row = 1)
text = Entry(root,width =50)
text.grid(column = 1,row = 1)

#Logica do programa
            
def excel():
    n = 0
    _sheet = _inputSheet.get()
    _text = text.get()
    df =pd.read_excel(f"C:/Users/start/Downloads/{_sheet}.xlsx")
    alunos = df["Aluno(a)"]
    telefones = df["Telefone(s)"]
    num = len(telefones)
    
    nav = webdriver.Chrome()
    nav.get('https://web.whatsapp.com/')

    while len(nav.find_elements_by_id('side')) < 1:
        time.sleep(1)

    while n <= len(telefones): 
        link =f"https://web.whatsapp.com/send?phone={telefones[n]}&text={_text}"
        nav.get(link)
        n+= 1



#Buttons
button = Button(root,bg="blue",text="Mandar",command=excel)
button.grid(column = 1 ,row=2)


#Second buttons
_nameShee= Label(root,text ="Nome Planilha  :  ")
_nameShee.grid(column = 0,row = 3)
_inputShee = Entry(root,width = 50)
_inputShee.grid(column = 1,row = 3)

text_mensage = Label(root,text ="Mensagem  :  ")
text_mensage.grid(column = 0,row = 4)
texte = Entry(root,width =50)
texte.grid(column = 1,row = 4)

root.mainloop()
```
